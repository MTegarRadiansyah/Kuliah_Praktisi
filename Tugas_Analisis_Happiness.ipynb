## Saat ini kita akan Membandingkan Happiness faktor dari Negara Bahrain dengan Greece

```python
# Filter data untuk negara Bahrain dan Yunani
bahrain_data = df[df['Country'] == 'Bahrain'].iloc[0]
greece_data = df[df['Country'] == 'Greece'].iloc[0]

# Columns yang akan diplot pada grafik radar
columns = ['Dystopia (1.83) + residual', 'Explained by: GDP per capita',
           'Explained by: Social support', 'Explained by: Healthy life expectancy',
           'Explained by: Freedom to make life choices',
           'Explained by: Generosity', 'Explained by: Perceptions of corruption']

# Values untuk Bahrain dan Yunani
bahrain_values = [bahrain_data[col] for col in columns]
greece_values = [greece_data[col] for col in columns]

# Jumlah variabel yang akan diplot
num_vars = len(columns)

# Menghitung sudut untuk setiap variabel
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Menyempurnakan loop

# Membuat plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plot untuk Bahrain
ax.fill(angles, bahrain_values + [bahrain_values[0]], color='blue', alpha=0.4, label='Bahrain')
ax.plot(angles, bahrain_values + [bahrain_values[0]], color='blue', linewidth=2)

# Plot untuk Yunani
ax.fill(angles, greece_values + [greece_values[0]], color='green', alpha=0.4, label='Greece')
ax.plot(angles, greece_values + [greece_values[0]], color='green', linewidth=2)

# Label untuk setiap titik
ax.set_xticks(angles[:-1])
ax.set_xticklabels(columns)

# Judul
plt.title('Happiness Factors: Bahrain vs. Greece', size=15, color='black', y=1.1)

# Menampilkan legenda
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

plt.show()
```

```python
# Filter data untuk Finlandia
finland_data = df[df['Country'] == 'Finland'].iloc[0]

# Kolom-kolom yang akan diplot
columns = ['Explained by: GDP per capita',
           'Explained by: Social support',
           'Explained by: Healthy life expectancy',
           'Explained by: Freedom to make life choices',
           'Explained by: Generosity',
           'Explained by: Perceptions of corruption']

# Nilai untuk Finlandia
finland_values = [finland_data[col] for col in columns]

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(columns, finland_values, color='skyblue')
plt.xlabel('Contribution to Happiness Score', fontsize=12)
plt.ylabel('Factors', fontsize=12)
plt.title('Contribution of Factors to Finland\'s Happiness Score', fontsize=14)
plt.gca().invert_yaxis()  # Supaya faktor teratas di plot terletak di bagian atas
plt.show()
```
